```python
import numpy as np
import scipy as sc
import math
from numpy import linalg
from numpy.linalg import inv
import random
from numpy.linalg import solve
```

```python
# create random A of size n x n
n = 100
matrix = np.zeros(shape=(n,n))

for i in range(n):
    for j in range(n):
        matrix[i][j] = random.randint(-100, 100)
```

```python
#LU decomp (code from part a)

lower = np.zeros(shape=(n, n))
upper = np.zeros(shape=(n, n))

# Performs LU decomposition
for i in range(n):
    # U matrix
    for k in range(i, n):
        sum = 0
        for j in range(0, i):
            sum += lower[i][j]*upper[j][k]
        upper[i][k] = matrix[i][k] - sum

    # L matrix
    for k in range(i, n):
        if i == k:
            lower[i][i] = 1
        else:
            sum = 0
            for j in range(i):
                sum += lower[k][j]*upper[j][i]
            lower[k][i] = (matrix[k][i] - sum) / upper[i][i]
```

```python
# create random b of size n x 1

b = np.zeros(shape=(n, 1))

for i in range(n):
    b[i][0] = random.randint(-100, 100)
```

```python
# evaluate using inverse (i)

inverse_A = inv(matrix)
x_inverseMethod = np.matmul(inverse_A, b)
```

```python
# evaluate using built in funct (ii)
x_builtIn = solve(matrix, b)
```

```python
# evaluate using LU (iii)
y = solve(lower, b)
x_LU = solve(upper, y)
```

```python
# check that they match

print(x_LU)
print(x_inverseMethod)
print(x_builtIn)
```
